```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = ['male', 'female', 'male', 'female', 'male', 'female', 'male', 'female', 'male', 'female', 'male', 'female']
y = ['small', 'med', 'large', 'small', 'med', 'large', 'small', 'med', 'large', 'small', 'med', 'large']
z = [0, 1, 1, 0, 1, 0, 1, 0, 1, 1, 0, 1]

# Create a dictionary to count occurrences of each combination
counts = {}
for i in range(len(x)):
    key = (x[i], y[i], z[i])
    counts[key] = counts.get(key, 0) + 1

# Create a list of unique x and y values
x_labels = sorted(list(set(x)))
y_labels = sorted(list(set(y)))

# Create a 2D array to store the counts
heatmap_data = np.zeros((len(y_labels), len(x_labels)))

# Populate the 2D array with counts
for i, y_val in enumerate(y_labels):
    for j, x_val in enumerate(x_labels):
        for z_val in [0, 1]:  # Consider both z-values
            key = (x_val, y_val, z_val)
            heatmap_data[i, j] += counts.get(key, 0)

# Create the heatmap
fig, ax = plt.subplots()
im = ax.imshow(heatmap_data, cmap='YlOrRd')  # Choose an appropriate colormap

# Add colorbar
cbar = ax.figure.colorbar(im, ax=ax)
cbar.ax.set_ylabel('Count', rotation=-90, va="bottom")

# Set ticks and labels
ax.set_xticks(np.arange(len(x_labels)))
ax.set_yticks(np.arange(len(y_labels)))
ax.set_xticklabels(x_labels)
ax.set_yticklabels(y_labels)

# Rotate x-axis labels
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Add annotations with counts
for i in range(len(y_labels)):
    for j in range(len(x_labels)):
        text = ax.text(j, i, int(heatmap_data[i, j]),
                       ha="center", va="center", color="black")

# Add title
ax.set_title("Heatmap of Gender, Size, and Z-value")

# Display the heatmap
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Sample data (same as before)
x = ['male', 'female', 'male', 'female', 'male', 'female', 'male', 'female', 'male', 'female', 'male', 'female'] * 2
y = ['small', 'med', 'large', 'small', 'med', 'large', 'small', 'med', 'large', 'small', 'med', 'large'] * 2
z = [0, 1, 1, 0, 1, 0, 1, 0, 1, 1, 0, 1] * 2

# Create a pandas DataFrame
df = pd.DataFrame({'gender': x, 'size': y, 'result': z})

# Create a pivot table to count occurrences, separating by 'result'
heatmap_data_0 = df[df['result'] == 0].pivot_table(index='size', columns='gender', values='result', aggfunc='count', fill_value=0)
heatmap_data_1 = df[df['result'] == 1].pivot_table(index='size', columns='gender', values='result', aggfunc='count', fill_value=0)

# Create the heatmaps
fig, axes = plt.subplots(1, 2, figsize=(12, 5))  # 1 row, 2 columns for two heatmaps

# Heatmap for result = 0
im0 = axes[0].imshow(heatmap_data_0, cmap='Blues')
axes[0].set_title("Heatmap of Gender, Size, and Result = 0")

# Heatmap for result = 1
im1 = axes[1].imshow(heatmap_data_1, cmap='Reds')
axes[1].set_title("Heatmap of Gender, Size, and Result = 1")

# Add colorbars
cbar0 = fig.colorbar(im0, ax=axes[0])
cbar0.ax.set_ylabel('Count', rotation=-90, va="bottom")
cbar1 = fig.colorbar(im1, ax=axes[1])
cbar1.ax.set_ylabel('Count', rotation=-90, va="bottom")

# Set ticks and labels for both heatmaps
for i, ax in enumerate(axes):
    ax.set_xticks(np.arange(len(heatmap_data_0.columns)))
    ax.set_yticks(np.arange(len(heatmap_data_0.index)))
    ax.set_xticklabels(heatmap_data_0.columns)
    ax.set_yticklabels(heatmap_data_0.index)
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

    # Add annotations with counts
    for i in range(len(heatmap_data_0.index)):
        for j in range(len(heatmap_data_0.columns)):
            if i == 0:
                heatmap_data = heatmap_data_0
            else:
                heatmap_data = heatmap_data_1
            text = ax.text(j, i, int(heatmap_data.iloc[i, j]),
                           ha="center", va="center", color="black")

# Display the heatmaps
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Sample data (same as before)
x = ['male', 'female', 'male', 'female', 'male', 'female', 'male', 'female', 'male', 'female', 'male', 'female'] * 2
y = ['small', 'med', 'large', 'small', 'med', 'large', 'small', 'med', 'large', 'small', 'med', 'large'] * 2
z = [0, 1, 1, 0, 1, 0, 1, 0, 1, 1, 0, 1] * 2

# Create a pandas DataFrame
df = pd.DataFrame({'gender': x, 'size': y, 'result': z})

# Create a pivot table to count occurrences
heatmap_data = df.pivot_table(index='size', columns='gender', values='result', aggfunc='count', fill_value=0)

# Create the heatmap
fig, ax = plt.subplots(figsize=(8, 6))
im = ax.imshow(heatmap_data, cmap='YlOrRd', alpha=0.5)  # Use alpha for transparency

# Add colorbar
cbar = ax.figure.colorbar(im, ax=ax)
cbar.ax.set_ylabel('Count', rotation=-90, va="bottom")

# Set ticks and labels
ax.set_xticks(np.arange(len(heatmap_data.columns)))
ax.set_yticks(np.arange(len(heatmap_data.index)))
ax.set_xticklabels(heatmap_data.columns)
ax.set_yticklabels(heatmap_data.index)

# Rotate x-axis labels
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# --- Add scatter points ---
# Create separate DataFrames for result = 0 and result = 1
df_0 = df[df['result'] == 0]
df_1 = df[df['result'] == 1]

# Get unique combinations and their counts for each result value
comb_counts_0 = df_0.groupby(['gender', 'size']).size().reset_index(name='counts')
comb_counts_1 = df_1.groupby(['gender', 'size']).size().reset_index(name='counts')

# Scatter plot for result = 0 (blue)
ax.scatter(comb_counts_0['gender'].map({'male': 0, 'female': 1}),
           comb_counts_0['size'].map({'small': 0, 'med': 1, 'large': 2}),
           s=comb_counts_0['counts'] * 50,  # Scale size based on count
           c='blue', label='Result = 0', alpha=0.7)

# Scatter plot for result = 1 (red)
ax.scatter(comb_counts_1['gender'].map({'male': 0, 'female': 1}),
           comb_counts_1['size'].map({'small': 0, 'med': 1, 'large': 2}),
           s=comb_counts_1['counts'] * 50,  # Scale size based on count
           c='red', label='Result = 1', alpha=0.7)

# Add legend
ax.legend()

# Add title
ax.set_title("Heatmap with Scatter Points for Gender, Size, and Result")

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
x = ['male', 'female', 'male', 'female', 'male', 'female', 'male', 'female', 'male', 'female', 'male', 'female'] * 2
y = ['small', 'med', 'large', 'small', 'med', 'large', 'small', 'med', 'large', 'small', 'med', 'large'] * 2
z = [0, 1, 1, 0, 1, 0, 1, 0, 1, 1, 0, 1] * 2

# Create a pandas DataFrame
df = pd.DataFrame({'gender': x, 'size': y, 'result': z})

# Group by 'gender', 'size', and 'result' to count occurrences
counts = df.groupby(['gender', 'size', 'result']).size().reset_index(name='counts')

# Create the scatter plot with scaled dot size and colored by 'result'
plt.figure(figsize=(8, 6))

# Scatter plot for result = 0 (blue)
plt.scatter(counts[counts['result'] == 0]['gender'],
            counts[counts['result'] == 0]['size'],
            s=counts[counts['result'] == 0]['counts'] * 50,  # Scale size based on count
            c='blue', label='Result = 0', alpha=0.7)

# Scatter plot for result = 1 (red)
plt.scatter(counts[counts['result'] == 1]['gender'],
            counts[counts['result'] == 1]['size'],
            s=counts[counts['result'] == 1]['counts'] * 50,  # Scale size based on count
            c='red', label='Result = 1', alpha=0.7)

# Add labels and title
plt.xlabel('Gender')
plt.ylabel('Size')
plt.title('Scatter Plot with Scaled Dot Size and Color by Result')
plt.legend()

# Show the plot
plt.show()
```
